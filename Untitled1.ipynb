```python
import pandas as pd

df = pd.read_csv("student.csv")   # change to your dataset name
df
print(df.head())   # Top 5 rows
print(df.tail())   # Bottom 5 rows
df.info() # Show Info
df.describe() # Stats
X = df[["Math" , "Reading"]]   # Features
Y = df[["Writing"]]                # Label
print(X.head())
print(Y.head())
```

```python
import numpy as np

# ---- Step 1: Set number of features (d) and samples (n) ----
d = 3   # number of features
n = 5   # number of samples

# ---- Step 2: Create Weight Vector W ∈ R^d ----
W = np.random.rand(d, 1)   # column vector (d × 1)

# ---- Step 3: Create Feature Matrix X ∈ R^(d×n) ----
X = np.random.rand(d, n)   # matrix (d × n)

# ---- Step 4: Compute Y = W^T X ----
Y = W.T @ X   # result will be (1 × n)

# ---- Step 5: Print Results ----
print("Weight Vector W:\n", W)
print("\nFeature Matrix X:\n", X)
print("\nOutput Vector Y = W^T X:\n", Y)
```
